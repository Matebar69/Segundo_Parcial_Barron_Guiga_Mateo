```python
from Bio import SeqIO
from Bio.Seq import Seq

gen_res1 = open("blatem1.fasta")
leer_gen_res1 = gen_res1.read()
x= leer_gen_res1[84:]
print("Secuencia del primer genoma resistente (blatem1.fasta): ")
print(x)
print("_" * 60)
print(" ")
gen_res2 = open("blactxm15.fasta")
leer_gen_res2 = gen_res2.read()
y= leer_gen_res2[161:]
print("Secuencia del segundo genoma resistente (blactxm15.fasta): ")
print(y)
print("_" * 60)
print(" ")
def cargar_gen_resistencia(path_fasta):
    for record in SeqIO.parse(path_fasta, "fasta"):
        return str(record.seq)


def cargar_genoma(path_fna):
    secuencias = []
    for record in SeqIO.parse(path_fna, "fasta"):
        secuencias.append(str(record.seq))
    return secuencias

def buscar_gen_res1(gen_fasta, archivo_fna):
    gen_resistencia = cargar_gen_resistencia(gen_fasta)
    gen_reverso = str(Seq(gen_resistencia).reverse_complement())
    
    secuencias_genoma = cargar_genoma(archivo_fna)
    
    for i, sec in enumerate(secuencias_genoma):
        if gen_resistencia in sec:
            print(f"Gen encontrado en la secuencia {i+1} en orientación directa.")
            return True
        if gen_reverso in sec:
            print(f"Gen encontrado en la secuencia {i+1} en orientación reversa complementaria.")
            return True
    print(f"{gen_fasta} NO encontrado en el archivo {archivo_fna}")
    return 

buscar_gen_res1("blatem1.fasta", "GCF_050389795.1_MS10495_ARL02201_ST410_genomic.fna")
buscar_gen_res1("blatem1.fasta", "GCF_050516345.1_ASM5051634v1_genomic.fna")
buscar_gen_res1("blatem1.fasta", "GCF_050516355.1_ASM5051635v1_genomic.fna")
buscar_gen_res1("blatem1.fasta", "GCF_050516365.1_ASM5051636v1_genomic.fna")
buscar_gen_res1("blatem1.fasta", "GCF_050516375.1_ASM5051637v1_genomic.fna")
buscar_gen_res1("blatem1.fasta", "GCF_050516385.1_ASM5051638v1_genomic.fna")
buscar_gen_res1("blatem1.fasta", "GCF_050563505.1_ASM5056350v1_genomic.fna")
buscar_gen_res1("blatem1.fasta", "GCF_050568845.1_2848_genomic.fna")
print("_"*100)
print("")
buscar_gen_res1("blactxm15.fasta", "GCF_050389795.1_MS10495_ARL02201_ST410_genomic.fna")
buscar_gen_res1("blactxm15.fasta", "GCF_050516345.1_ASM5051634v1_genomic.fna")
buscar_gen_res1("blactxm15.fasta", "GCF_050516355.1_ASM5051635v1_genomic.fna")
buscar_gen_res1("blactxm15.fasta", "GCF_050516365.1_ASM5051636v1_genomic.fna")
buscar_gen_res1("blactxm15.fasta", "GCF_050516375.1_ASM5051637v1_genomic.fna")
buscar_gen_res1("blactxm15.fasta", "GCF_050516385.1_ASM5051638v1_genomic.fna")
buscar_gen_res1("blactxm15.fasta", "GCF_050563505.1_ASM5056350v1_genomic.fna")
buscar_gen_res1("blactxm15.fasta", "GCF_050568845.1_2848_genomic.fna")




```
